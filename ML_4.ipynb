```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import sklearn
```

**1. Чтение данных**

```python
url = 'https://raw.githubusercontent.com/nastosimys/Machine_learning/refs/heads/main/diabetes_dataset.csv'
data_raw = pd.read_csv(url)
data_raw.replace("No Info", np.nan, inplace=True)
data_raw.replace("Other", np.nan, inplace=True)
```

```python
data_raw
```

```python
data_raw['gender'] = data_raw['gender'].astype('category')
data_raw['location'] = data_raw['location'].astype('category')
data_raw['smoking_history'] = data_raw['smoking_history'].astype('category')
```

**2. Обработка пропущенных значений**

```python
data_raw.isnull()
```

Количество пропущенных значений:

```python
data_raw.isnull().sum()
```

Названия столбцов, в которых есть пропущенные значения:

```python
[feature for feature in data_raw.columns if data_raw[feature].isnull().sum() > 0]
```

```python
from sklearn.impute import SimpleImputer
imp_cat = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_cat.fit(data_raw[['gender', 'smoking_history']])
```

```python
data_raw[['gender', 'smoking_history']] = imp_cat.transform(data_raw[['gender', 'smoking_history']])
```

Теперь можно убедиться, что пропущенных значений нет:

```python
data_raw.isna().sum()
```

**3. Обработка категориальных признаков**

```python
data_raw['gender'] = data_raw['gender'].map({'Female': 1, 'Male': 0})
data_raw['gender']
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(drop='if_binary', sparse_output=False)
data_raw_smoking = pd.DataFrame(data_raw['smoking_history'])
enc.fit(data_raw_smoking)
```

```python
dummies = pd.DataFrame(enc.transform(data_raw_smoking),
                       columns=enc.get_feature_names_out(), index=data_raw.index)
dummies.head()
```

```python
data_raw = pd.concat((data_raw, dummies), axis=1).drop('smoking_history', axis=1)
data_raw = data_raw.rename(columns={'smoking_history_not current': 'smoking_history_not_current'})
data_raw
```

**4. Нормализация**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
norm_data = data_raw[['year', 'age', 'bmi', 'hbA1c_level', 'blood_glucose_level']]
scaler = StandardScaler()
scaler.fit(norm_data)
norm_data = pd.DataFrame(scaler.transform(norm_data), columns=norm_data.columns, index=norm_data.index)
```

```python
norm_data
```

```python
data_raw = data_raw.drop(['year', 'age', 'bmi', 'hbA1c_level', 'blood_glucose_level'], axis=1)
data_raw = pd.concat((data_raw, norm_data), axis=1)
data_raw
```

**5. Решение задачи кластеризации**

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn import metrics
from sklearn.metrics import adjusted_rand_score, mutual_info_score
from scipy.stats import entropy
```

```python
X = data_raw.drop(['diabetes', 'year', 'location'], axis=1)
y = data_raw['diabetes']
```

**KMeans**

```python
k = 10
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)
cluster_labels = kmeans.labels_
silhouette_score_kmeans = metrics.silhouette_score(X, cluster_labels)
ari_kmeans = adjusted_rand_score(y, cluster_labels)
mi = mutual_info_score(y, cluster_labels)
h_y = entropy(y)
h_c = entropy(cluster_labels)
nmi_kmeans = (2 * mi) / (h_y + h_c)

print('Силуэтный коэффициент:', silhouette_score_kmeans)
print('ARI:', ari_kmeans)
print('NMI:', nmi_kmeans)
```

**DBSCAN**

```python
dbscan = DBSCAN(eps=1.5, min_samples=10)
dbscan.fit(X)
dbscan_labels = dbscan.labels_
silhouette_score_dbscan = metrics.silhouette_score(X, dbscan_labels)
ari_dbscan = adjusted_rand_score(y, dbscan_labels)
mi = mutual_info_score(y, dbscan_labels)
h_y = entropy(y)
h_c = entropy(dbscan_labels)
nmi_dbscan = (2 * mi) / (h_y + h_c)

print('Силуэтный коэффициент:', silhouette_score_dbscan)
print('ARI:', ari_dbscan)
print('NMI:', nmi_dbscan)
```

**GMM**

```python
gm = GaussianMixture(n_components=4)
gm.fit(X)
silhouette_score_gm = metrics.silhouette_score(X, gm.predict(X))
ari_gm = adjusted_rand_score(y, gm.predict(X))
mi = mutual_info_score(y, gm.predict(X))
h_y = entropy(y)
h_c = entropy(gm.predict(X))
nmi_gm = (2 * mi) / (h_y + h_c)

print('Силуэтный коэффициент:', silhouette_score_gm)
print('ARI:', ari_gm)
print('NMI:', nmi_gm)
```
